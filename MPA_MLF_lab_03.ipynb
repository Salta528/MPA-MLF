# SVM for classification

Load IRIS dataset, check its contents:


```python
from sklearn.datasets import load_iris
iris=load_iris()
iris.feature_names
print(iris.feature_names)
print(iris.data[0:5,:])
print(iris.target[0:5])
print(iris.data)
```

Split data into training and testing parts:

```python
from sklearn.model_selection import train_test_split
X=iris.data
y=iris.target
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
print(X.shape)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
print(y_train)
print(y_test)
```

Use a Support Vector Machine for classification:


```python
from sklearn.svm import SVC
SVMmodel=SVC(kernel='linear')
SVMmodel.fit(X_train,y_train)
print(SVMmodel.get_params())
SVMmodel.score(X_test,y_test)
```

Let's explore more now.

*   Choose only first two features (columns) of iris.data
*   SVM is in its basic form a 2-class classifier, so eliminate iris.target =2 from the data



```python
X=iris.data[:,0:2]
y=iris.target
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

Plot scatterplots of targets 0 and 1 and check the separability of the classes:

```python
import matplotlib.pyplot as plt
plt.scatter(X[:,0], X[:,1])
plt.scatter(X_train[y_train==0,0], X_train[y_train==0,1], color='red')
plt.scatter(X_train[y_train==1,0], X_train[y_train==1,1], color='blue')
plt.scatter(X_train[y_train==2,0], X_train[y_train==2,1], color='green')
plt.show()
```

Train and test the SVM classifier, play with regularization parameter C (either use the default value or try e.g. 200)

```python
X=iris.data[iris.target!=2,0:2] # Select only the first two features and exclude class '2' from the Iris dataset
print(X.shape)
y=iris.target[iris.target!=2] # Create target variable, only including classes 0 and 1
SVMmodel_1=SVC(kernel='linear', C=200) # Initialize an SVM classifier
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2) # Split dataset into 80% training and 20% testing
SVMmodel_1.fit(X_train,y_train)
SVMmodel_1.score(X_test,y_test)
```

Show support vectors in the 2D plot, plot the decision line from equation [w0 w1]*[x0 x1] + b = 0:

```python
supvectors=SVMmodel_1.support_vectors_
# Plot the support vectors here

#Separating line coefficients:
# Use SVMmodel_1 which was trained on 2 classes
W=SVMmodel_1.coef_
b=SVMmodel_1.intercept_
print(W)
print(b)
import numpy as np
x1=np.linspace(np.min(X[:,0]),np.max(X[:,0]),200)
# Calculate x2 using the correct dimensions of W and b
x2=-b/W[0,1]-W[0,0]/W[0,1]*x1
plt.scatter(X[y==0,0], X[y==0,1], color='red')
plt.scatter(X[y==1,0], X[y==1,1], color='blue')
plt.scatter(x1,x2, color='black')
```

# Anomaly detection via SVM

Import one-class SVM and generate data (Gaussian blobs in 2D-plane):

```python
from sklearn.svm import OneClassSVM
from sklearn.datasets import make_blobs
from numpy import quantile, where, random

random.seed(11)
x, _ = make_blobs(n_samples=300, centers=1, cluster_std=.3, center_box=(4, 4)) # Generate a dataset with 300 samples, 1 cluster, and standard deviation of 0.3; the cluster is centered around (4,4)

plt.scatter(x[:,0], x[:,1]) # Plot the generated datase
plt.show()

```

Train one-class SVM and plot the outliers (outputs of prediction being equal to -1)

```python
SVMmodelOne = OneClassSVM(kernel='rbf', gamma=0.001, nu=0.03)


SVMmodelOne.fit(x)
pred = SVMmodelOne.predict(x) # Predict anomalies
print(pred)
anom_index = where(pred==-1) # Identify indices of outliers where prediction == -1
values = x[anom_index]
print(anom_index)  # Print indices
print(values)  # Print coordinates
plt.scatter(x[:,0], x[:,1]) # Scatter plot of all data points
plt.scatter(values[:,0], values[:,1], color='red') #highlight in red
plt.axis('equal')
plt.legend()
plt.show()

```

Plot the support vectors:

What if we want to have a control what is outlier? Use e.g. 5% "quantile" to mark the outliers. Every point with lower score than threshold will be an outlier.

```python
scores = SVMmodelOne.score_samples(x)

thresh = np.quantile(scores, 0.05) # Set threshold as the 5% quantile
print(thresh)
index = np.where(scores<=thresh) # Find indices where scores are below the outlier
values = x[index] # Extract outlier values from dataset

plt.scatter(x[:,0], x[:,1], label="Normal Data", color="blue") # Scatter plot of all data points
plt.scatter(values[:,0], values[:,1],label="Outliers", color='red')
plt.axis('equal')
plt.legend()
plt.show()
```
